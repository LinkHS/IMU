Fourier transform using `numpy.fft` method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# How many time points are needed i,e., Sampling Frequency
samplingFrequency = 100

# At what intervals time points are sampled
samplingInterval = 1 / samplingFrequency

# Begin/End time period of the signals
beginTime = 0
endTime = 10

# Frequency of the signals
signal1Frequency = 4
signal2Frequency = 7

# Time points
time = np.arange(beginTime, endTime, samplingInterval)
```

```python
# Create two sine waves
amplitude1 = np.sin(2*np.pi*signal1Frequency*time)
amplitude2 = np.sin(2*np.pi*signal2Frequency*time)

# Create subplot
figure, axis = plt.subplots(2, 1)
plt.subplots_adjust(hspace=1)

axis[0].plot(time, amplitude1)
axis[1].plot(time, amplitude2)

for ax, fq in zip(axis, (4, 7)):
    ax.set_title(f'Sine wave with a frequency of {fq} Hz')
    ax.set_xlabel('Time')
    ax.set_ylabel('Amplitude')

plt.show()
```

```python
# Add the sine waves
amplitude = amplitude1 + amplitude2

# Time domain representation of the resultant sine wave
plt.plot(time, amplitude)
plt.title('Sine wave with multiple frequencies')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Frequency domain representation
# Normalize amplitude
fourierTransform = np.fft.fft(amplitude)/len(amplitude)

# Exclude sampling frequency
fourierTransform = fourierTransform[range(int(len(amplitude)/2))]

tpCount = len(amplitude)
values = np.arange(int(tpCount/2))
timePeriod = tpCount/samplingFrequency
frequencies = values/timePeriod

# Frequency domain representation
plt.title('Fourier transform depicting the frequency components')
plt.plot(frequencies, abs(fourierTransform))
plt.xlabel('Frequency')
plt.ylabel('Amplitude')

plt.show()
```

```python
freq = np.fft.rfftfreq(amplitude.size, samplingInterval)
ap = np.fft.rfft(amplitude)
plt.plot(freq, abs(ap))
plt.show()
```
